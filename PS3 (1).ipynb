## 2. Explore Central Limit Theorem

```python
# Function that calculates the expected value of user inputted list using probability
def expected_value(values, weights):
    values = np.asarray(values)
    weights = np.asarray(weights)
    return (values * weights).sum() / weights.sum()
first = np.random.randint(0,2, size = 10)*2 - 1 # Generates the Bernoulli-distributed numbers
first_prob = [0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5] # The probability of the random numbers

expected = (expected_value(first, first_prob))
print("Expected value is: ", expected)
# Function that calculates the variance of user inputted list
def variance(user):
    avg = sum(user) / len(user)
    var1 = sum((x-avg)**2 for x in user) / len(user)
    return var1
print("Variance is : ", variance(first))
```
